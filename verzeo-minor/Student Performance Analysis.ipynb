```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data_file = "StudentsPerformance.csv"
df = pd.read_csv(data_file)
```

```python
df['average'] = (df['math score'] + df['reading score'] + df['writing score'])/3
df.head(10)
```

**LUNCH**
----------
The following plot clearly shows that children who recieve lunch at standard prices tend to get more marks on an average as compared to those who recieve lunch at reduced prices.  
Further on an average it is observed that females get more marks than males!

```python
sns.catplot(x = 'lunch', y ='average', hue = 'gender',kind = 'violin', data = df)
```

**TEST PRERATION COURSE**
---------------------------
the boxplot shows that children that have completed the **preparation course** have achived more marks on an average than people who havent completed the preparation course.  
Further plots show that the preparation course helped students score more in all the subjects

```python
sns.catplot(x = 'test preparation course', y ='average',kind = 'box', data = df)
```

```python
fig, axes = plt.subplots(1,3, figsize = (10,5))
fig.tight_layout(w_pad = 3.0, h_pad = 3.0)
sns.boxplot('test preparation course', y ='math score', data = df, ax = axes[0])
sns.boxplot('test preparation course', y ='reading score', data = df, ax = axes[1])
sns.boxplot('test preparation course', y ='writing score', data = df, ax = axes[2])
plt.show()
```

**Parental Level of Education**
----------------------------------------------
It can be seen from the box plot that children whose parents have master's degree get a highest average marks.  
**the average marks are keep in increasing as the parental education increases**

```python
chart = sns.catplot(x = 'parental level of education', y ='average',kind = 'box', data = df )
chart.set_xticklabels(rotation=40, ha="right")
```

**Race/Ethnicity**
------------------------
children from group A get the least average marks while children from group E get the maximum average marks

```python
sns.catplot(x = 'race/ethnicity', y ='average', kind = 'box', data = df)
```

**Relation between reading and writing scores**
-------------------------
From the relplot we can see that the reading and writing scores are more or less propotional in a linear fashion.  
Also the female data points lie mostly above the male datapoints throughout the dataset.

```python
sns.relplot(x='reading score', y='writing score', kind ='scatter', hue = 'gender', data=df)
```

**How Girls and Boys Perform in various Subjects**
----------------------
Females seem to have a higher average score in reading and writing skills but Males do better than females in Mathematics.  
For average score Females have a higher average than Males!

```python
fig, axes = plt.subplots(2, 2, figsize = (10,10))
fig.tight_layout(w_pad = 3.0, h_pad = 3.0)
sns.violinplot(x='gender', y = 'math score', data = df, ax = axes[0,0])
sns.violinplot(x='gender', y = 'reading score', data = df, ax = axes[0,1])
sns.violinplot(x='gender', y = 'writing score', data = df, ax = axes[1,0])
sns.violinplot(x='gender', y = 'average', data = df, ax = axes[1,1])
plt.show()
```
