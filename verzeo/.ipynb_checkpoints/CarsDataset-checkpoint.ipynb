## Data Imputation
Since the data contains a lot of missing values we need to either remove those rows or fill them using a statistical apprach.
Lets take a look at how many rows have missing values for each column.

```python
print("Before Imputation:")
print("Rows missing Mileage: ",df[df['Mileage'].isnull()].shape[0])
print("Rows missing Engine: ",df[df['Engine'].isnull()].shape[0])
print("Rows missing Power: ",df[df['Power'].isnull()].shape[0])
print("Rows missing Seats: ",df[df['Seats'].isnull()].shape[0])
print("Total Rows: ",df.shape[0])
```

### Seats
Seats can be imputed by looking at what car the data is about. Since similar cars have same number of seats we can use the value of seats that are available to fill in the missing valued for the cars whose values are not available.

first I prepare a dictionary mapping car names to the number of seats in them. Then I have implemented a fuzzy search algorithm that can match substrings in two different strings i.e (''Mariti Suzuki Alto' matches 'Maruti Suzuki Also LXI') and then fill in the missing value from the row it matches with.

### Engine
The same logic is applicable for engines, since same cars have the same type of engine we can compare the missing value with other rows that have the same car name.

### Mileage
Mileage is only missing 2 rows so we can just remove them withouth replacing the values

```python
seats_dict = {}
for num in np.unique(df[df['Seats'].notnull()].Seats):
    seats_dict[num] = np.unique(df[df['Seats'] == num].Model)

engine_dict = {}
for e in np.unique(df[df['Engine'].notnull()].Engine):
    engine_dict[e] = np.unique(df[df['Engine'] == e].Model)

def get_Seats(row):
    for num, cars in seats_dict.items():
        found = False
        for i in [model.find(row.Model) for model in cars]:
            if i != -1:
                found = True
                break
        if found:
            return num

def get_Engines(row):
    for e, cars in engine_dict.items():
        found = False
        for i in [model.find(row.Model) for model in cars]:
            if i != -1:
                found = True
                break
        if found:
            return e

df['Seats'] = df.apply(lambda row: get_Seats(row) if np.isnan(row.Seats) else row.Seats, axis=1)
df['Engine'] = df.apply(lambda row: get_Engines(row) if np.isnan(row.Engine) else row.Engine, axis=1)
```

#### There are still a lot of rows that were not filled so in the following cell we will remove them from the dataset

```python
df = df[df['Mileage'].notnull()]
df = df[df['Seats'].notnull()]
df = df[df['Engine'].notnull()]
df = df.reset_index(drop=True)
```

## Data Imputation with KMeans
The **Power** column has the most number of missing rows so we need to impute them and cannot jus simply remove them. It is reasonable to think that Power, Mileage and Engine are closely related.
Here we will use KMeans clusturing to cluster the cars based on **Mileage** and **Engine** and then impute the missing values ass the mean values of that cluster

```python
kmeans_df = df[['Mileage','Engine']]
kmeans = KMeans(n_clusters = 5, max_iter=1000)
kmeans.fit(kmeans_df)
```

### Finding the optimum number of clusters
In order to successfully use KMeans we need to first find the the optimal number of cluster we should divide the data into. 

In order to do this we will use the **Elbow Method**. Basically I plotted the **Squared Sum Erroe (SSE)** for a few clusters and plot a graph. The graph will show a elbow like structure. Now the optimum number of clusters is the point at which elbow is formed.

```python
kmax = 8

def calculate_sse(kmax):
    sse = []
    for i in range(1, kmax+1):
        kmeans = KMeans(n_clusters = i, max_iter=1000).fit(kmeans_df)
        centroids =  kmeans.cluster_centers_
        pred_clusters = kmeans.predict(kmeans_df)
        curr_sse = 0
        
        for index,row in kmeans_df.iterrows():
            curr_center = centroids[pred_clusters[index]]
            curr_sse += (row[0]-curr_center[0])**2 + (row[1]-curr_center[1])**2
    
        sse.append(curr_sse)
    
    return sse
        
sse = calculate_sse(kmax)
plt.plot(sse)
```

In the above graph we can clearly see that the elbow occurs at K = 3 or K = 4. Here I will use K = 4
We will cluster the data in to 4 clusters and the for each missing values will will predict its cluster and then fill the value with the mean value for that cluster

```python
optim_clusters = 4

kmeans = KMeans(n_clusters = optim_clusters).fit(kmeans_df)
df['cluster'] = kmeans.predict(kmeans_df)

average_dict = {}
for i in range(optim_clusters):
    average_dict[i] = df[df.cluster == i]['Power'].mean()
    
df['Power'] = df.apply(lambda row: average_dict[row.cluster] if np.isnan(row.Power) else row.Power, axis = 1)

df.drop(columns = ['cluster'], inplace=True)
del(kmeans_df)
```

## Outlier Removal
So now we will remove the Outliers from the dataset using the Z-Score method. We will calulate the z-score for all the rows and then remove all the rows with higher z-score using a threshold value (here I used 3)

```python
fig, axis = plt.subplots(2,3, figsize = (15,10))
sns.boxplot(df['Kilometers_Driven'], ax=axis[0,0])
sns.boxplot(df['Mileage'], ax=axis[0,1])
sns.boxplot(df['Price'], ax=axis[0,2])
sns.boxplot(df['Power'], ax=axis[1,0])
sns.boxplot(df['Engine'], ax=axis[1,1])
```

```python
def remove_outlier(column, df, threshold):
    bool_df = df.apply(lambda row: abs((row[column] - df[column].mean())/df[column].std(ddof=0)) < threshold, axis = 1)
    df = df[bool_df]
    return df

df_o = remove_outlier('Kilometers_Driven', df, 2)
df_o = remove_outlier('Mileage', df_o, 3)
df_o = remove_outlier('Price', df_o, 3)
df_o = remove_outlier('Power', df_o, 3)
df_o = remove_outlier('Engine', df_o, 3)
```

## After Outlier Removal
we can see that after outlier removal the Box Plots show relatively less outliers as before.
